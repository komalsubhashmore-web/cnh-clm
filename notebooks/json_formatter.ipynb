```python
import json

# Set your filename here
filename = "generated_redteam_dataset_gemini.jsonl"

# Open and read the file
with open(filename, "r") as f:
    lines = f.readlines()

# Format and display
for idx, line in enumerate(lines, 1):
    try:
        data = json.loads(line)
        print(f"\nExample {idx}")
        print("=" * 80)
        print("Instruction:")
        print(data["instruction"].strip())
        print("\nOutput:")
        print(data["output"].strip())
        print("=" * 80)
    except Exception as e:
        print(f"\n[!] Error parsing line {idx}: {e}")
```
