## Feed images to model that extracts the text and creates a new document

```python
import os
import re
import shutil
import pandas as pd
from urllib.parse import unquote
```

```python

'''
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    if os.path.isfile(file_path):
        print("Fil:", filename)'''

def process_gitbook_images(input_path, output_path, image_texts_folder,missing_images):

    with open(input_path, 'r', encoding='utf-8') as infile:
        lines = infile.readlines()

    # Matches. ![](<../.gitbook/assets/diagram1.png>)
    image_pattern = re.compile( r'../../.gitbook/assets/(.*?)(.png|.jpg|.jpeg|.pdf)')

    output_lines = []

    for line in lines:
        match = image_pattern.search(line)
        if match: 
            image_name = match.group(1)
            #print(f"\n Image found!")
            #print(f"- Filename: {image_name}")

            # Make paranteses (, ) to  underscore _
            image_name_original = image_name
            image_name = unquote(image_name)
            image_name = image_name.replace("(", "_").replace(")", "_")
            image_name = image_name.replace("-", "_")
            
            #image_name_file = 'image1'  # Change to image_name later
            
            # Collect text from external folder
            image_path = f"{image_texts_folder}/{image_name}.txt"
            # Put the missing images into txt file 
            missing_log_path = "missing_images.txt"  
            
            if os.path.exists(image_path):
                with open(image_path, "r", encoding="utf-8") as f:
                    text_from_image = f.read()
                    #print(text_from_image)
                    # Skriv ny rad med titel och beskrivning
                new_line = f"\n{text_from_image}\n"
                output_lines.append(new_line)
            else:
                print(f"{image_name} not found. \n Line: {line} \n \n")

                ########################################################
                # Save the lost images:
                ########################################################
                possible_extensions = [".jpg", ".jpeg", ".png", ".bmp", ".gif"]
                source_folder = "images"
                destination_folder = "lost_images/"
                os.makedirs(destination_folder, exist_ok=True)
                found = False
                for ext in possible_extensions:
                    full_path = os.path.join(source_folder, image_name + ext)
                    if os.path.exists(full_path):
                        shutil.copy(full_path, os.path.join(destination_folder, image_name + ext))
                        #print(f"Copied {full_path} to {destination_folder}")
                        found = True
                        break

                if not found:
                    print("Image file not found with any of the common extensions.")

            
                missing_images.append({
                    "missing_image": image_name,
                    "markdown_file": input_path  # make sure this variable is defined in your loop
                })
            

        else:
            output_lines.append(line)
    os.makedirs(os.path.dirname(output_path), exist_ok=True)
    filename = f'{output_path}'
    with open(filename, 'w', encoding='utf-8') as outfile:
        outfile.writelines(output_lines)

    #print(f"\n New markdown file saved as: {filename}")

def create_missing_list(listname):
    # Initialize list to collect missing entries
    listname = []
    return listname

def save_list_csv(csv_list,rootfolder):
    # Create DataFrame from missing entries
    df = pd.DataFrame(csv_list)

    # Save to CSV or view it
    df.to_csv(f"missing_{rootfolder}.csv", index=False)

    
### Change root folder here:
FOLDER_PATH = 'miscellaneous-reversing-forensics' #'offensive-security' #'offensive-security' # 'offensive-security-experiments' # 'miscellaneous-reversing-forensics'

count = 0

missing_images = create_missing_list('missing_images')
# root(path to root), dirs(list with subfolders), files(list with filenames)
for root, dirs, files in os.walk(FOLDER_PATH):
    for filename in files:
        if filename.endswith(".md"):
            file_path = os.path.join(root, filename)
            #print(file_path)
            #print(filename)   
            process_gitbook_images(file_path, f'Converted/{file_path}', 'ocr_outputs', missing_images)
save_list_csv(missing_images, FOLDER_PATH)

### Test 
#input = 'offensive-security-experiments/active-directory-kerberos-abuse/active-directory-lab-with-hyper-v-and-powershell.md'    ### Use in the loop where filepath is here
#updated_markdown = 'updated.md'    ### Use in the loop where updated_markdown = filename ###
#textfolder = 'test_texts'              ### Change to the real folder later ###
#process_gitbook_images(input, f"test_outputs/{updated_markdown}", textfolder )
           
```
